```python
import requests
url = 'http://api.citybik.es/v2/networks/bicicorunha'
```

```python
#Consulta os slots que hay valeiros nas 5 primeirs estacion mais o nome de cada unha.
response = requests.get(url)
for estacion in range(0,5):
    slots=response.json()['network']['stations'][estacion]['empty_slots']
    nombre=response.json()['network']['stations'][estacion]['name']
    print(f'Na estación {nombre} hay {slots} bicicletas usandose')
```

```python
#Consulta as bicicletas que hay dispoñibles nas 5 ultimas  estacions mais o nome de cada unha.
response = requests.get(url)
for estacion in range(17,22):
    slots=response.json()['network']['stations'][estacion]['free_bikes']
    nombre=response.json()['network']['stations'][estacion]['name']
    print(f'Na estación {nombre} hay {slots} bicicletas dispoñibles')
```

```python
#Consulta as cordenadas da estacion atopada na Estación de Autobuses.
response = requests.get(url)
for estacion in range(0,22):
    s=response.json()['network']['stations'][estacion]['name']

    if s=='Estación Autobuses':
          la=response.json()['network']['stations'][estacion]['latitude']
          lo=response.json()['network']['stations'][estacion]['longitude']
          print(f'Las coredenadas de Estación Autobuses son: latitud {la}y longitud {lo}.')
```

```python
#Crea un Dataframe cos nomes das estacións mais os slots valeiros mais as biciletas que estan libres
import pandas as pd
response = requests.get(url)
lista=[]
for estacion in range(0,22):
    slots_libres=response.json()['network']['stations'][estacion]['empty_slots']
    slots_ocupados=response.json()['network']['stations'][estacion]['free_bikes']
    nombre=response.json()['network']['stations'][estacion]['name']
    lista.append([nombre,slots_ocupados,slots_libres])
estaciones=pd.DataFrame(lista,columns=['Estación','Bicicletas Libres','Bicicletas Ocupadas'])
estaciones=estaciones.set_index('Estación')
estaciones
```

```python
#Por ultimo engade ao Dataframe antereior o estado de cada estación
for estacion in range(0,22):
    estado=response.json()['network']['stations'][estacion]['extra']['status']
    lista[estacion].append(estado)
estaciones=pd.DataFrame(lista,columns=['Estación','Bicicletas Libres','Bicicletas Ocupadas','Estado'])
estaciones=estaciones.set_index('Estación')
estaciones
```
